```python
#Import the dependencies
import pandas as pd
from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import requests
```

```python
url = 'https://www.leadingrestaurants.co.uk/data/full-lists/hardens/'

# Retrieve pages with request
response = requests.get(url)

response.text
```

```python
# Create BS object; parse with html.parser
soup = bs(response.text, 'html.parser')
type(soup)
```

```python
# Examine the results
print(soup.prettify())
```

```python
# Print all paragraph texts
paragraphs = soup.find_all('a')

for paragraph in paragraphs:
    print(paragraph.text)
    print("----")
```

```python
def create_column(data, column_name: str, results):
    data[column_name] = results
    return data

def find_div(class_name, soup=soup):
    return soup.find_all('div', class_=class_name)

def create_list(results):
    return [res.text.strip("\n").replace("\n", "").strip(" ") for res in results]

data = pd.DataFrame()

class_names = ["res_name", "res_location", "top_pos"]
columns = ["restaurants", "location", "rank"]

for class_name, column in zip(class_names, columns):
    results = find_div(class_name)
    results = create_list(results)
    data = create_column(data, column, results)
    
data
```
